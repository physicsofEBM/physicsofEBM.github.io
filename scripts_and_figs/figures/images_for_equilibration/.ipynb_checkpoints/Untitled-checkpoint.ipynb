```python
%matplotlib inline
import matplotlib.pyplot as plt
url = "zero/resized_damaged_zeros_visible_0.jpg"

im = plt.imread(url)
plt.imshow(im)
```

```python
im_new = (im/100>1)*1
plt.imshow(im_new)
```

```python
im_flat = im_new.reshape(100,)
print(im_flat.shape)
```

```python
liste = []
for s in range(100):
    x = s%10
    y = np.floor(s/10)
    c = im_flat[s]
    liste.append([x,y,c])   
```

Translate to JSON

```python
import json

json.dumps(np.array(liste).tolist())
```

### Make loop for all images

```python
folder = "four/"
number = "fours"
v_or_h = "visible"

list_one_nr = []
for index in range(8):
    url = folder + "resized_damaged_"+number+"_"+v_or_h+"_"+str(index)+".jpg"
    im = plt.imread(url)
    im_new = (im/100>1)*1
    im_flat = im_new.reshape(100,) 
    list_one_nr.append(im_flat)
json.dumps(np.array(list_one_nr).tolist())    
```
